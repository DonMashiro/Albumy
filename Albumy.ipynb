```python
import pandas as pd
```

```python
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
```

```python
type(data)
```

```python
df=data[0]
```

```python
#Zamień nagłówki kolumn na polskie odpowiedniki.
df.rename(columns={'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':'ROK','HIGH POSN':'MAX POZ'},inplace=True)
```

```python
print(df.columns)
```

```python
#Ilu pojedynczych artystów znajduje się na liście?
liczba_pojedynczych_artystowt=df['ARTYSTA'].nunique()
print(f"Na liście znajduje się {liczba_pojedynczych_artystow} pojedynczych artystów.")
```

```python
#Które zespoły pojawiają się najczęściej na liście?
najczesciej_występujący_artysta=df['ARTYSTA'].value_counts().idxmax()
print(f"Najczęściej występującym zespołem na liście jest {najczesciej_występujący_artysta}.")
```

```python
#Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.
df.rename(str.capitalize, axis='columns',inplace=True)
```

```python
print(df.columns)
```

```python
#Wyrzuć z tabeli kolumnę ‘Max Poz’.
df.drop('Max poz',axis=1)
```

```python
#W którym roku wyszło najwięcej albumów znajdujących się na liście?
rok_z_najwieksza_liczba_albumow = df['Rok'].value_counts().idxmax()
print(f"Najwięcej albumów wyszło w roku {rok_z_najwieksza_liczba_albumow}.")
```

```python
#Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?
liczba_albumow_wydanych=df['Rok'].between(1960, 1990).sum()
print(f"Na liście znajduje się {liczba_albumow_wydanych} albumów wydanych między 1960 a 1990 rokiem włącznie.")
```

```python
#W którym roku wydany został najmłodszy album na liście?
najmlodszy_album=df['Rok'].max()
print(f"Najmłodszy album wydany został w {najmlodszy_album} roku.")
```

```python
df.groupby(['Artysta','Rok']).min()
```

```python
df.drop_duplicates(subset=['Artysta'], keep='first')
```
